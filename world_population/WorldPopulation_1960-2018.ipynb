# World Population 1960 - 2018

```python
import numpy as np
import pandas as pd
from pathlib import Path

# Import
df = pd.read_csv(Path('data/world-population-19602018/population_total_long.csv'))

# Head
df.head()
```

## Selection

```python
country_selection = df['Country Name'] == 'Germany'

selected_country = df.loc[country_selection]

selected_country_2000 = selected_country.loc[df['Year']>=2000]
```

## Delta

```python
df = selected_country_2000.copy()

df = df.reset_index(drop=True)

df['Diff'] = df['Count'].diff()

df['Delta']=(df['Count'] - df['Count'].shift(1))

df
```
