# English Word Frequency

English word frequency

Data from: https://www.kaggle.com/rtatman/english-word-frequency

```python
import numpy as np
import pandas as pd
from pathlib import Path

# Import
df = pd.read_csv(Path('./data/unigram_freq.csv'))

# Head
df.head()
```

## Top 100 - used english words

```python
dataset = df.loc[~df['word'].isin(['c', 'e', 's', 'x', 'n', 'b', 't', 'd', 'm', 'r', 'p', 'f', 'l', 'w', 'o', 'g', 'h', 'v', 'k', 'y', 'j', 'u', 'q', 'z'])]
dataset = dataset.sort_values(by='count', ascending=False)

dataset.reset_index()

dataset = dataset[:100]


print("Position: Word (Count)")

i = 0
for index, data in dataset.iterrows():
    word, count = data
    i = i + 1
    print(i,": ", word, " (", count, ")")

```
